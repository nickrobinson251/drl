```python
from valkyrie.envs.two_room_grid_env import TwoRoomGridEnv
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
env = TwoRoomGridEnv()
env.render()
```

```python
for _ in range(30):
    action = env.action_space.sample()
    obs, reward, done, info = env.step(action)
    if action != 0 and info['success']:
        env.render()
    if done:
        break
print(env.episode_total_reward)
```
